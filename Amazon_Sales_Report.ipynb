```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/ds/Amazon Sale Report.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['New','PendingS'], axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
df['ship-postal-code']=df['ship-postal-code'].astype(int)
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.rename(columns={'Qty':'Quantity'})
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Size',data=df)

for bars in ax.containers:
  ax.bar_label(bars)
```

Из графика выше видно, что размер M самый популярный


```python
df.groupby(['Size'], as_index = False)['Qty'].sum().sort_values(by='Qty', ascending=False)
```

```python
dist_qty = df.groupby(['Size'], as_index = False)['Qty'].sum().sort_values(by='Qty', ascending=False)

ax = sns.barplot(x='Size', y='Qty', data=dist_qty)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))

ax=sns.countplot(data=df, x='Courier Status',hue= 'Status')
```

Из графика видно, что большинство заказов отправляются курьером

```python
sns.histplot(data=df, x=df['Size'])
```

```python
df['Category'] = df['Category'].astype(str)
```

```python
column_data = df['Category']
plt.figure(figsize=(10, 5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90);
```

Чаще всего покупают футболки

```python
B2B_Check = df['B2B'].value_counts()

plt.pie(B2B_Check, labels=B2B_Check, autopct='%1.1f%%');
```

99.2% - ретейлеры, 0.8% - b2b

```python

x_data = df['Category']
y_data = df['Size']


sns.scatterplot(data = df, x=x_data,y=y_data);
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python

top_10_state = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

Покупателей из Maharashtra больше всех
